#### Q1: Можно ли доверять результатам анализа? Почему?

Можно, рассмотрим предположения F-теста:
1. Наблюдения приходится считать независимыми
2. По левым центральному и нижнему графикам остатков видно, что они распределены нормально
3. По левому верхнему графику диагностики видно, что остатки образуют равномерную (не расширяющуюся и не сужающуюся) область, что говорит об их гомоскедостичности (одинаковости дисперсий)

#### Q2: При условии, что уровень значимости равен 0.05, является ли значимым общий F-тест для модели? Почему?
...

Является, так как значение F-value значительно превышает 1

Помимо этого, значительно малое P-value говорит о том, что следует отвергнуть гипотезу о эквивалетности эффетивности мастеров на разных брендах

#### Q3: Стоит ли удалить пересечение из модели? Почему?
...

Нет, так как пересечение увеличивает объясненную вариативность (R-square)
