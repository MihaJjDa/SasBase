```sas
/* Ваш код для выполнения задания (не забудьте запустить его) */
proc reg data = &DS rmse;
    ALL_REG: model &ORIG_TARGET = &inputs 
    / selection = rquare adjrsq cp stop=5;
    ods output SubsetSelSummary = SubSelSum;
run; 

proc sql noprint;
    select VarsInModel
    into :model2
    from SubSelSum
    where RMSE = (select min(RMSE)
                  from SubSelSum);
run;

proc reg data = &DS;
    model &ORIG_TARGET = &model2;
run;

quit;
```
